#**Data cleaning**
Робота із пропущеними даними, некоректними даними, аномаліями.



## Перевірка на наявність пропущених значень у кожній таблиці, оцінюю їх долю, оцінюю причини їх виникнення.
Заповнюю або видаляю пропущені значення.

```python
#кількість пропущених значень в products
print(products.isna().sum())
```

```python
#кількість пропущених значень в countries
print(countries.isna().sum())
```

```python
#доля пропущенних значень
print(countries.isna().sum() / countries.shape[0] * 100)
```

```python
#перевіряємо, в яких рядках є пропущені дані.
missing_rows = countries[countries.isna().any(axis=1)]
print(missing_rows)
```

```python
#змінюю пропущені дані на логічно вірні
countries["region"].fillna("Antarctica", inplace=True)
countries["sub-region"].fillna("Antarctica", inplace=True)
countries["alpha-2"].fillna("NA", inplace=True)

print(countries.isna().sum())
```

```python
#перейменуємо назви стовпців events таким чином, щоб вони не містили великих літер, і були у вигляді snake_case
events.columns = events.columns.str.lower().str.replace(" ", "_")
events.info()
```

```python
#кількість пропущених значень в events
print(events.isna().sum())
#доля пропущенних значень
print(events.isna().sum() / events.shape[0] * 100)
```

```python
#перевіряємо, в яких рядках є пропущені дані.
missing_rows = events[events.isna().any(axis=1)]
print(missing_rows)
```

```python
#змінюю пропущене значення на загальне "Unknown"
events["country_code"].fillna("Unknown", inplace=True)
# видаляю рядки, що мають пропущені значення в Units Sold
events.dropna(subset=["units_sold"], inplace=True)
print(events.isna().sum())
```

## Перевірка, чи всі типи даних розпізнаються коректно. Є розбіжності, тому роблю перетворення типів у колонках.

```python
#Order Date та   Ship Date мають невірний формат object, потрібно змінити на datetime
events["order_date"] = pd.to_datetime(events["order_date"], errors='coerce')

print(events["order_date"].dtypes)

events["ship_date"] = pd.to_datetime(events["ship_date"])
print(events["ship_date"].dtypes)
events.info()
```

##З’єднання трьох таблиць в один датафрейм. Видаляю непотрібні для аналізу колонки. Роблю перейменування колонок.

```python
# перейменування  колонки id в датафреймі products на product_id

products.rename(columns={"id": "product_id"}, inplace=True)
products.info()
```

```python
# з’єднання events та products в один датафрейм та
left_join = pd.merge(events, products, on="product_id", how="left")
print(left_join)
```

```python
# перейменування  колонки alpha-3 в датафреймі countries на country_code
countries.rename(columns={"alpha-3": "country_code"}, inplace=True)
countries.rename(columns={"name": "country_name"}, inplace=True)
countries.info()
```

```python
# з’єднання events, products та countries в один датафрейм
full_data = pd.merge(left_join, countries, on="country_code", how="left")
print(left_join)
```

```python
full_data.info()
```
